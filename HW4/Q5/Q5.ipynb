```python
# Normal BFS + DFS non-recursive
from collections import defaultdict 

class Graph: 

    def __init__(self,Number): 
        self.graph = defaultdict(list) 
        self.numberOfNodes = Number
        self.bfs=[]
        self.dfs=[]
    def addEdge(self,u,v,w): 
        if v not in self.graph[u]:
            self.graph[u].append(v) 
            self.graph[v].append(u)
  
    def BFS(self, s): 
  

        visited = [False] * (self.numberOfNodes) 

        queue = [] 

        queue.append(s) 
        visited[s] = True
  
        while len(queue)!=0: 
  
            s = queue.pop(0) 
            self.bfs.append(s)
        
            for i in self.graph[s]: 
                if visited[i] == False: 
                    queue.append(i) 
                    visited[i] = True
#         print(self.bfs)
        
  
    def DFS(self,s): 
        visited = [False] *(self.numberOfNodes)
 
        stack = [] 
 
        stack.append(s)  
  
        while (len(stack)):  
 
            s =  stack.pop() 

            if (not visited[s]):  
                self.dfs.append(s)
#                 print(s,end=' ') 
                visited[s] = True 
    
            for node in self.graph[s]:  
                if (not visited[node]):  
                    stack.append(node)  
            
#         print(self.dfs)
        
        
        
# g = Graph(9) 
# g.addEdge(0, 7, 8) 
# g.addEdge(0, 1, 4) 
# g.addEdge(1, 7, 11)
# g.addEdge(1, 2, 8)
# g.addEdge(2, 3, 7)
# g.addEdge(2, 8, 2)
# g.addEdge(3, 4, 9)
# g.addEdge(3, 5, 14)
# g.addEdge(4, 5, 10)
# g.addEdge(5, 6, 2)
# g.addEdge(2,5, 4)
# g.addEdge(6, 8, 6)
# g.addEdge(6, 7, 1)
# g.addEdge(7, 8, 7)
# print(g.graph)
# g.DFS(0)
```

```python
f = open("NYC.txt","r")
x = f.readlines()
number_of_nodes = int(x[0])
number_of_edges = int(x[1])
new_york = Graph(number_of_nodes)
for i in range(number_of_edges):
    a = x[2+i].split()
    u = int(a[0])
    v = int(a[1])
    w = float(a[2])
    new_york.addEdge(u,v,w)
```
